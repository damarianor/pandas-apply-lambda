##  Challenge 1. Warm up. New column

```python
def movie_votes(votes):
    if 0 < votes <= 999:
        return 1
    elif 1000 < votes <= 9999:
        return 2
    elif 10000 < votes <= 99999:
        return 3
    elif 100000 < votes <= 999999:
        return 4
    else:
        return 5
```

```python
df['bins'] = df.apply(lambda x: movie_votes( x['Votes']), axis = 1)
```

We want to know how much is the revenue per minute for every movie.

## Challenge 3. Using the lambda

We want to create a new ranking where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.

```python
def new_ranking(genre):
    if 'Thriller' in genre:
        return 1
    elif 'Comedy' in genre:
        return -1
```

```python
df['New Ranking'] = df.apply(lambda x: new_ranking( x['Genre']), axis = 1)
```

## Challenge 4. Now the real stuff

We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers.

```python
def ascii_sum(word, votes):
    a = sum(ord(char) for char in word)
    b =  int(a / votes)
    if b < 2:  
        return False
    else:
        return True
```

```python
df['Ascii values'] = df.apply(lambda x: ascii_sum( x["Title"], x["Votes"]), axis=1)
```
